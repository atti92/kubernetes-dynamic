```python
import kubernetes_dynamic as kd


class SP(kd.models.ResourceValue):
    description: str
    
class AssemblyElements(kd.models.ResourceValue):
    sp: SP

class AssemblySpec(kd.models.ResourceValue):
    elements: AssemblyElements

class Assembly(kd.models.ResourceItem):
    name: str
    spec: AssemblySpec


cl = kd.K8sClient()

asm = cl.get_api(kind="Assembly", object_type=Assembly).get(name="kodbc-assembly")
asm
```
